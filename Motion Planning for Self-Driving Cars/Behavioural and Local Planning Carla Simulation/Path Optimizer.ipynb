```python
from path_optimizer import PathOptimizer
import matplotlib.pyplot as plt
import numpy as np
```

```python
xf, yf, tf = 2, -1, 1  # Adjust these as desired for different spirals
optimizer = PathOptimizer()
optimized_spiral = optimizer.optimize_spiral(xf, yf, tf)

# For comparison, generate a non-optimized spiral
# You could set random parameters, or use the initial guess used in the optimizer for a more direct comparison.  
non_optimized_spiral = optimizer.sample_spiral([0.0, 0.0, np.linalg.norm([xf, yf])]) 

plt.figure(figsize=(10, 6))

# Plot the optimized spiral
plt.plot(optimized_spiral[0], optimized_spiral[1], 'b-', label='Optimized Spiral')
# Plot the non-optimized spiral
plt.plot(non_optimized_spiral[0], non_optimized_spiral[1], 'r-', label='Non-Optimized Spiral')

# Mark the start and end points
plt.plot(0, 0, 'go', label='Start')
plt.plot(xf, yf, 'ro', label='Target End')

# Enhance the plot
plt.title("Path Optimization Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensure equal aspect ratio
plt.show()
```
